```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('nTop ASME Hackathon Data.csv')
df.head(1)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Velocity Inlet'], df['Mass'], alpha=0.7)
plt.xlabel('Velocity Inlet')
plt.ylabel('Mass')
plt.title('Scatter plot between Velocity Inlet and Mass')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['X Cell Size'], df['PressureDrop'], alpha=0.7)
plt.xlabel('X Cell Size')
plt.ylabel('Mass')
plt.title('Scatter plot between X Cell Size and Mass')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap Between All Columns')
```
